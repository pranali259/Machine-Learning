# Skewness

```python
df.describe()
```

```python
from scipy.stats import skew
```

```python
df[['OverallQual','OverallCond']] = df[['OverallQual','OverallCond']].astype(str)
house_num=df.select_dtypes(["int64","float64"])
```

```python
house_num.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
house_num
```

```python
house_num.set_index("Id",inplace=True)
```

```python
house_num
```

```python
for col in house_num:
    print(col,skew(house_num[col]))
```

```python
for col in house_num:
    print(col,skew(house_num[col]))
    plt.figure()
    sns.distplot(house_num[col])
    plt.plot()
```

```python
for col in house_num:
    if skew(house_num[col])>1 or skew(house_num[col])<-1:
        house_num[col]=np.sqrt(house_num[col])
```

```python
for col in house_num:
    print(col,skew(house_num[col]))
    plt.figure()
    sns.distplot(house_num[col])
    plt.plot()
```

```python
for col in house_num:
    plt.figure
    sns.boxplot(data=house_num, x=col,whis=3)
    plt.show()
```
