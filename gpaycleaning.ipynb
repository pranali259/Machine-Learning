```python
df["Price"]=df["Price"].str.replace("$",'',regex=True)
```

```python
df=df.drop(10472)
```

```python
def func(x):
    if "k" in x:
        x=x.replace("k","")
        return float(x)/1024
    elif "M" in x:
        x=x.replace("M","")
        return float(x)
    elif x=="Varies with device":
        return np.nan
    else:
        return float(x)
```

```python
func("200k")
```

```python
df['Size']=df['Size'].map(func)
```

```python
df["Size"].unique()
```

```python
df.head()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```
